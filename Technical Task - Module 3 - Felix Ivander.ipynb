#### Q1: What features contributes the most to the progress of diabetes quantity?
#### A1:

```python
from sklearn.ensemble import RandomForestRegressor
# define dataset
X= diabetes_df.drop(['dis_progress'], axis=1)
y= diabetes_df['dis_progress']

# define the model
model = RandomForestRegressor()
# fit the model
model.fit(X, y)
# get importance
importance = model.feature_importances_
# summarize feature importance
for i,v in enumerate(importance):
    print(f'Feature: {i}, Score: {v:.5f}')
# plot feature importance
plt.bar(list(X.columns), importance)
plt.show()
```

Through the visualization above, we can conclude the importance of each feature.
The most important important features that affect the diabetes progression are:
* bmi
* bp
* ltg

#### Q2: Does aging accelerate the disease progression?
#### A2:

```python
plt.figure(figsize=(16,9))
sns.regplot(data=diabetes_df , x='age', y='dis_progress', line_kws={"color": "red"})
plt.show()
```

Based on the data, we can see that the older someone gets, it slightly affects the diabetes progression positively

#### Q3: How does BMI affect the disease progression?
#### A3:

```python
diabetes_df['bmi'].value_counts()
```

```python
plt.figure(figsize=(16,9))
sns.regplot(data=diabetes_df , x='bmi', y='dis_progress', line_kws={"color": "red"})
plt.show()
```

Based on the data and the graph shown above, BMI is strongly positively correlated with diabetes progression.

The higher the bmi, the higher the diabetes progression becomes.

#### Q4: Does the amount of low density lipoproteins & high density lipoproteins accelerate/slows the progress of diabetes?
#### A4:

```python
plt.figure(figsize=(9,5))
sns.regplot(data=diabetes_df , x='ldl', y='dis_progress', line_kws={"color": "red"})
plt.show()
```

```python
plt.figure(figsize=(9,5))
sns.regplot(data=diabetes_df , x='hdl', y='dis_progress', line_kws={"color": "red"})
plt.show()
```

Through both graphs we can see that:
* ldl (Low Density Lipoproteins) have a weak positive corellation with the diabetes progression
* hdl (High Density Lipoproteins) have a strong negative corellation with the diabetes progression

#### Q5: Does higher blood sugar level lead to bigger progress of diabetes?
#### A5:

```python
plt.figure(figsize=(9,5))
sns.regplot(data=diabetes_df , x='glu', y='dis_progress', line_kws={"color": "red"})
plt.show()
```
